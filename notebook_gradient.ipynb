```python
import matplotlib.pyplot as plt
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D


```

```python
def test_parameters(model, aa, ww, X):
    y = X[0, -500:]
    x = np.arange(0., len(y))
    z = np.ndarray(shape=(len(ww), len(aa)))
    for j, a in enumerate(aa):
        for i, w in enumerate(ww):
            model.model.freeze_parameters([[1, 1, 1], [1, 1, 1]],
                                   [[1., 0.28284271, 3.14159265],
                                    [a, w, 0.]])
            
            z[i, j] = model.model.loss(model.model(x), y)
    return z
```

```python
axis_a = np.arange(0.1, 2.5, 0.02)
axis_w = np.arange(-0.15, 0.15, 0.001)
a, w = np.meshgrid(axis_a, axis_w)
z = test_parameters(model, axis_a, axis_w, X[0:1,:])
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.view_init(20, 40)
surf = ax.plot_surface(a, w, z, cmap=cm.coolwarm,
                       linewidth=0, antialiased=False)
fig.colorbar(surf, shrink=0.5, aspect=5)
plt.gcf().set_size_inches(20.5, 10.5)

plt.show()
```
